```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
RiskRate = pd.read_csv('C:\Work Folders\Risk data\Risk_1.csv',encoding='iso-8859-1')
```

```python
RiskRate.head()
```

```python
pearsoncorr = RiskRate.corr(method='pearson')
```

```python
pearsoncorr
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Risk factors")
sb.heatmap(pearsoncorr, ax=ax, annot=True,linewidth=.5)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
fig.tight_layout()
plt.show()
```
